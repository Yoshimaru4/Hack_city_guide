```python
import os
import pandas as pd

# Define the path to the folder containing CSV files
folder_path = 'datasets/details_data'

# List all CSV files in the directory
csv_files = [file for file in os.listdir(folder_path) if file.endswith('.csv')]

# Initialize an empty DataFrame to hold combined data
combined_df = pd.DataFrame()

# Iterate over each CSV file and append its data
for file in csv_files:
    file_path = os.path.join(folder_path, file)
    try:
        df = pd.read_csv(file_path)  # Adjust 'sep' if necessary
        combined_df = pd.concat([combined_df, df], ignore_index=True)
        print(f"Successfully read {file}")
    except Exception as e:
        print(f"Error reading {file}: {e}")

# Check if 'Name' column exists and remove duplicates
if 'Name' in combined_df.columns:
    unique_df = combined_df.drop_duplicates(subset=['Name'])
    print(f"Total entries after removing duplicates: {len(unique_df)}")
else:
    print("The 'Name' column does not exist in the combined data.")
    unique_df = combined_df  # Proceed without removing duplicates

# Define the output file path
output_file = os.path.join(folder_path, 'combined_unique_data.csv')

# Save the unique combined data to the output CSV
unique_df.to_csv(output_file, index=False)  # Adjust 'sep' if necessary

print(f"Combined CSV file with unique 'Name' entries saved to {output_file}")
```

```python
from transformers import VitsModel, AutoTokenizer
import torch

model = VitsModel.from_pretrained("facebook/mms-tts-rus")
tokenizer = AutoTokenizer.from_pretrained("facebook/mms-tts-rus")

text = "Привет! Начнем с интересной истории о Раушской набережной. Вы знаете, почему она так называется?"
inputs = tokenizer(text, return_tensors="pt")

with torch.no_grad():
        output = model(**inputs).waveform
```

```python
from IPython.display import Audio

Audio(output, rate=model.config.sampling_rate)
```
